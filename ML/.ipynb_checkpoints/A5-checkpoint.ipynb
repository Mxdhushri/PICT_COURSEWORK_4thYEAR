```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('sales_data_sample.csv', encoding ="unicode_escape")
```

```python
df #understand the dataset, its coloums and all
```

```python
df['Discount Applied'] = np.random.randint(0,2,size=len(df))
df
```

```python
df = df[['QUANTITYORDERED', 'PRICEEACH', 'SALES','Discount Applied']] # points in multidimension
```

```python
from sklearn.cluster import KMeans

intertia = [] 

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k)# kmeans model/algo
    kmeans.fit(df) # we fit/train the entire dataframe
    intertia.append(kmeans.inertia_) #error stored , if clusters more error less
    #ie dist b/w median and any point is less
```

```python
import seaborn as sns
```

```python
sns.lineplot(x=range(1, 11), y=intertia, ) # elbow is formed
```

```python
# after plotting we conclude Optimal K = 4
kmeans = KMeans(n_clusters=4)
df['Cluster']  = kmeans.fit_predict(df) #cluster no. coloum is added
```

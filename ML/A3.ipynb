```python
df = pd.read_csv('Churn_Modelling.csv')
```

```python
df
```

```python
df['Number of Schemes applied'] = np.random.randint(1,11,size=len(df)) #add new coloumn
df
```

```python
df = df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)#unessesary stuff drop
```

```python
gender = pd.get_dummies(df['Gender'], drop_first=True) # get dummy of gender eg: m=1, f=0
geography = pd.get_dummies(df['Geography'], drop_first=True)
df = pd.concat([df, gender, geography], axis=1) # conct ie add new col to table
df = df.drop(['Geography', 'Gender'], axis=1)   # drop previous cols
df = df.dropna()
df
```

```python
X = df.drop(['Exited'], axis=1)
Y = df['Exited']
```

```python
from sklearn.preprocessing import StandardScaler # normalization method is used for scaling(getting b/w 0-1 val)
sc = StandardScaler() # mean =0 sd=1
X = sc.fit_transform(X) # we fit x cause x has various categories na, y is just exited ie a number (0 or 1)
X
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y)
```

```python
from keras.models import Sequential # sequential is a Neural Network model [ip,hidden,op]
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(activation='relu', units=6))#6 neurons #relu=activation function
model.add(Dense(activation='relu', units=6))# hidden layer
model.add(Dense(activation='sigmoid', units=1)) # s funtion gives probability of class
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, batch_size=10, epochs=50) # increse batch size for fast results
#epoch is number of iterations to reduce b/w y_train and y_pred
```

```python
y_pred = model.predict(x_test)
y_pred = (y_pred > 0.5)  
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred)) #finds accuracy on how much similar y_test and y_pred are 
```

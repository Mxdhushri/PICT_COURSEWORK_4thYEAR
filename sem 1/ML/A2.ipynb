```python
import pandas as pd # provides DataFrames for structured data
import numpy as np #numerical computations
```

```python
df = pd.read_csv('emails.csv')
```

```python
df['UniqueWords'] = (df.iloc[:, 1:3001] > 0).sum(axis=1) #creates one more column contatining no. of unique words in all 3001 cols
df
```

```python
X = df.drop(['Prediction','Email No.'], axis=1) #axis=1 ie cols
y = df['Prediction']   #binary classifiaction of whether spam orr not spam
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier() #variable=model
model.fit(x_train,y_train) #fitting always on training data
y_pred = model.predict(x_test) #rectangle with 4 boxes analogy , compare y_test and y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score

# # Sample actual labels and predicted labels (1 = spam, 0 = not spam)
# y_true = [0, 1, 0, 1, 1, 0, 1, 0, 0, 1]  # Actual labels
# y_pred = [0, 1, 0, 1, 0, 0, 1, 1, 0, 1]  # Predicted labels

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

# Extracting TP, FP, FN, TN
TN, FP, FN, TP = cm.ravel()

# Calculate Precision and Recall
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

# Output
print("Confusion Matrix:")
print(cm)

print("\nIndividual metrics:")
print(f"True Positives (TP): {TP}")
print(f"False Positives (FP): {FP}")
print(f"False Negatives (FN): {FN}")
print(f"True Negatives (TN): {TN}")

print("\nCalculated Metrics:")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
```

```python
# Calculate metrics
Recall = TP / (TP + FN) if (TP + FN) > 0 else 0
Precision = TP / (TP + FP) if (TP + FP) > 0 else 0
Accuracy = (TP + TN) / (TP + FP + TN + FN) if (TP + FP + TN + FN) > 0 else 0

print("Precision:",Precision)
print("Recall:",Recall)
print("Accuracy:",Accuracy)
```
